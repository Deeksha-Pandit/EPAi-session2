````
from faker import Faker
from collections import namedtuple
import datetime
from datetime import datetime
fake = Faker()
````

````

def fake_profile():
    '''
    This function is used to generate 10000 profiles using the faker module in python
    It stores all this data in namedtuple
    '''
    p = []
    start = datetime.now()
    faker_profile = namedtuple('faker_profile',fake.profile().keys())
    for i in range(10000):
        p.append(faker_profile(**fake.profile()))
    end = datetime.now()
    #print(end-start)
    return p
fake_profile()
````

````
def calculate():
    '''
    This function calculates the maximum blood_group, average current location, oldest person age and average age
    '''
    import datetime
    from datetime import date
    group = {}
    loc_coordinates = [0,0]
    max_age = {'age': 0, 'profile': None}
    avg_age = 0
    calc_dob = ()
    date_today = datetime.date.today()
    bg=fake_profile()
    for i in bg:
        group[i.blood_group]= group.get(i.blood_group,0) + 1
        loc_coordinates[0]+=(i.current_location[0]/10000)
        loc_coordinates[1]+=(i.current_location[1]/10000)
        age = (date_today - i.birthdate).days
        if  age > max_age['age']:
            max_age['age'] = age
            max_age['profile'] = i
        avg_age += int(age/365)

    profiles = namedtuple('profiles', 'group loc_coordinates max_age avg_age')
    res = profiles(max(group), loc_coordinates, (max_age['profile'], int(max_age['age']/365)), avg_age/10000)
    return res
calculate()
````

````
def fake_profile_dict():
    '''
    This function is used to generate 10000 profiles using the faker module in python
    It stores all this data in dictionary
    '''
    d = dict()
    start = datetime.now()
    for i in range(10000):
        d[i]=(fake.profile())
    
    end = datetime.now()
    #print(end-start)
    return d
fake_profile_dict()
````

````
import random
def stocks():
    '''
    This function uses the faker module to generate 100 different company names
    Then we use namedtupes to create Stock info such as company name, symbol, opening maket rate, closing market rate and highest rate touched
    '''
    comp = []
    Stock = namedtuple('Stock', 'company_name symbol open high close')
    for i in range(100):
        company_name = fake.company()
        symbol = company_name+str(i)
        open = random.uniform(1000,10000)
        high = open * random.uniform(1,2)
        close = open * random.uniform(0.5,1)
        comp.append(Stock(company_name=company_name, symbol=symbol, open=open, high=high, close=close))
    return comp
stocks()
````
